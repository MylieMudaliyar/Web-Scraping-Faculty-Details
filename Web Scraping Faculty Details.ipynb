```python
import requests

from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# URL of the website to scrape
url = "https://www.stevens.edu/school-engineering-science/departments/mathematical-sciences/faculty"
```

```python
# Send an HTTP GET request to the website
response = requests.get(url)
```

```python
# Parse the HTML code using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# Extract the relevant information from the HTML code
details = []
#for row in soup.select('div.faculty-card_cc--faculty-card__Z2Pk4'): 
#     name = row.find('div', class_='faculty-card_name-title__By2JH').find('h3').get_text()
#     designation = row.find('div', class_='faculty-card_name-title__By2JH').find('h4').get_text()
#     email = row.find('div', class_='faculty-card_contact-email__tJfiV').find('a').get_text()
#     contact = row.find('div', class_='faculty-card_contact-phone__0Eh28').find('a').get_text()
#     office = row.find('div', class_='faculty-card_contact-office__Pqw03').text.strip()
    
for row in soup.select('div.faculty-card_cc--faculty-card__Z2Pk4'): 
    name_element = row.find('div', class_='faculty-card_name-title__By2JH').find('h3')
    designation_element = row.find('div', class_='faculty-card_name-title__By2JH').find('h4')
    email_element = row.find('div', class_='faculty-card_contact-email__tJfiV').find('a')
    
    # Additional check for contact element
    contact_parent_element = row.find('div', class_='faculty-card_contact-phone__0Eh28')
    contact_element = contact_parent_element.find('a') if contact_parent_element else None
   
    office_element = row.find('div', class_='faculty-card_contact-office__Pqw03')

    name = name_element.get_text() if name_element else "N/A"
    designation = designation_element.get_text() if designation_element else "N/A"
    #email = email_element.get_text() if email_element else "N/A"
    email = email_element.get_text().split(':', 1)[-1].strip() if email_element else "N/A"
    contact = contact_element.get_text() if contact_element else "N/A"
    office = office_element.get_text().strip() if office_element else "N/A"

    details.append([name, designation, email, contact, office])
```

```python
# Store the information in a pandas dataframe
df = pd.DataFrame(details, columns=['Name', 'Designation' , 'Email', 'Contact', 'Office'])
```

```python
# Add a delay between requests to avoid overwhelming the website with requests
time.sleep(1)
```

```python
# Export the data to a CSV file
df.to_csv('stevens-maths-faculty-details.csv', index=False)
```
